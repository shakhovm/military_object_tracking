```python
import numpy as np
import cv2
import time



def draw_flow(img, flow, step=16):

    h, w = img.shape[:2]
    y, x = np.mgrid[step/2:h:step, step/2:w:step].reshape(2,-1).astype(int)
    fx, fy = flow[y,x].T

    lines = np.vstack([x, y, x-fx, y-fy]).T.reshape(-1, 2, 2)
    lines = np.int32(lines + 0.5)

    img_bgr = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    cv2.polylines(img_bgr, lines, 0, (0, 255, 0))

    for (x1, y1), (_x2, _y2) in lines:
        cv2.circle(img_bgr, (x1, y1), 1, (0, 255, 0), -1)

    return img_bgr


def draw_hsv(flow):

    h, w = flow.shape[:2]
    fx, fy = flow[:,:,0], flow[:,:,1]

    ang = np.arctan2(fy, fx) + np.pi
    v = np.sqrt(fx*fx+fy*fy)

    hsv = np.zeros((h, w, 3), np.uint8)
    hsv[...,0] = ang*(180/np.pi/2)
    hsv[...,1] = 255
    hsv[...,2] = np.minimum(v*4, 255)
    bgr = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

    return bgr




cap = cv2.VideoCapture(0)

suc, prev = cap.read()
prevgray = cv2.cvtColor(prev, cv2.COLOR_BGR2GRAY)


while True:

    suc, img = cap.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # start time to calculate FPS
    start = time.time()


    flow = cv2.calcOpticalFlowFarneback(prevgray, gray, None, 0.5, 3, 15, 3, 5, 1.2, 0)
    
    prevgray = gray


    # End time
    end = time.time()
    # calculate the FPS for current frame detection
    fps = 1 / (end-start)

    print(f"{fps:.2f} FPS")

    cv2.imshow('flow', draw_flow(gray, flow))
    cv2.imshow('flow HSV', draw_hsv(flow))


    key = cv2.waitKey(5)
    if key == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```

```python
def wh2xy(bbox):
    return [bbox[0], bbox[1], bbox[2] + bbox[0], bbox[3] + bbox[1]]


def xy2wh(bbox):
    return [bbox[0], bbox[1], bbox[2] - bbox[0], bbox[3] - bbox[1]]


def int_wh2xy(bbox):
    return list(map(int, wh2xy(bbox)))


def int_xy2wh(bbox):
    return list(map(int, xy2wh(bbox)))
```

```python
# Module implements smoothing using cross-correlation.
import cv2
import numpy as np
from skimage.registration import phase_cross_correlation


def extract(rect_orig):
    """
    Function preprocess the image before smoothing.

    Arguments:
        rect_orig: np.array
    Returns:
        np.array
    """
    rect = cv2.cvtColor(rect_orig, cv2.COLOR_BGR2GRAY)
    sigma = 10
    ksize = 3
    rect = cv2.GaussianBlur(rect, (ksize, ksize), 0)
    # TODO draw contours and corr on the largest (hmm, bad idea, contours are too sensitive)
    return rect


def get_shift(prev_rect, current_rect_with_noise):
    """
    Function calculates the shift between two bounding boxes.

    Arguments:
        prev_rect: np.array
        current_rect_with_noise: np.array
    Returns:
        tuple(tuple, tuple)
    """
    rect_curr = extract(current_rect_with_noise)
    rect_prev = extract(prev_rect)

    rect_curr_f = cv2.flip(cv2.flip(rect_curr, 1), 0)
    rect_prev_f = cv2.flip(cv2.flip(rect_prev, 1), 0)

    mx_sh = np.max([rect_prev.shape, rect_curr.shape], 0)
    mx_sh_f = np.max([rect_prev_f.shape, rect_curr_f.shape], 0)

    # padding to one size
    rect_prev = np.pad(rect_prev, [(0, mx_sh[0] - rect_prev.shape[0]), (0, mx_sh[1] - rect_prev.shape[1])],
                       'constant')
    # print(np.where(rect_prev == 0))
    # m = rect_prev.copy()
    # m[m != 0] = 255
    
    # plt.imshow(m, cmap='gray')
    # plt.show()
    rect_curr = np.pad(rect_curr, [(0, mx_sh[0] - rect_curr.shape[0]), (0, mx_sh[1] - rect_curr.shape[1])],
                       'constant')
    rect_prev_f = np.pad(rect_prev_f,
                         [(0, mx_sh_f[0] - rect_prev_f.shape[0]), (0, mx_sh_f[1] - rect_prev_f.shape[1])],
                         'constant')
    rect_curr_f = np.pad(rect_curr_f,
                         [(0, mx_sh_f[0] - rect_curr_f.shape[0]), (0, mx_sh_f[1] - rect_curr_f.shape[1])],
                         'constant')

    # TODO - https://scikit-image.org/docs/stable/auto_examples/registration/plot_register_rotation.html#recover
    #  -rotation-and-scaling-differences-with-log-polar-transform
    shift, err0, phdiff = phase_cross_correlation(rect_prev, rect_curr, normalization="phase")
    shift_f, err1, phdiff = phase_cross_correlation(rect_prev_f, rect_curr_f, normalization="phase")
    return shift, shift_f


def extract_patch(image, bbox):
    """
    Function extracts patch from the image.

    Arguments:
        image: np.array
        bbox: np.array
    Returns:
        np.array
    """
    x, y, w, h = bbox
    return image[y:y + h, x:x + w]

def get_bbox(bbox1, bbox2):
    max_shape = np.max(np.vstack([bbox1, bbox2])[:,2:], axis=0)
    def new_bbox(bbox):
        bbox_w_diff, bbox_h_diff = bbox[2:] - max_shape
        return np.array([bbox[0] - (bbox_w_diff) // 2, 
                         bbox[1] - (bbox_h_diff) // 2, max_shape[0], max_shape[1]])
    return new_bbox(bbox1), new_bbox(bbox2)

def smooth_bbox(prev_rect, current_rect_with_noise, pred_bbox):
    """
    Function smooths a bounding box. Format of bboxes is - x y w h.

    Arguments:
        prev_rect: np.array
        current_rect_with_noise: np.array
        pred_bbox: np.array
    Returns:
        np.array
    """
    x0, y0, w, h = pred_bbox
    x1, y1 = x0 + w, y0 + h

    shift, shift_f = get_shift(prev_rect, current_rect_with_noise)
    # shift = np.clip(shift, -20, 20)
    # shift_f = np.clip(shift_f, -20, 20)
    # print(shift, shift_f)
    x0, x1, y0, y1 = int(x0 - shift[1]), int(x1 + shift_f[1]), int(y0 - shift[0]), int(y1 + shift_f[0])
    return [x0, y0, x1 - x0, y1 - y0], shift, shift_f

def clip_box(bbox, max_width, max_height):
    x0, y0, w, h = bbox
    x0 = np.clip(x0, 0, max_width)
    y0 = np.clip(y0, 0, max_height)
    x1 = np.clip(x0 + w, 0, max_width)
    y1 = np.clip(y0 + h, 0, max_height)
    return [x0, y0, x1 - x0, y1 - y0]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import fourier_shift
from scipy.fft import fft2, ifft2

def phase_correlation(img1, img2):
    # Compute the 2D FFTs of the images
    f_img1 = fft2(img1)
    f_img2 = fft2(img2)

    # Compute the cross-power spectrum
    cross_power_spectrum = np.conj(f_img1) * f_img2

    # Compute the phase correlation
    phase_correlation = ifft2(cross_power_spectrum / np.abs(cross_power_spectrum))

    # Find the peak in the phase correlation to get the translation
    shift = np.unravel_index(np.argmax(np.abs(phase_correlation)), phase_correlation.shape)

    return shift

def shift_image(img, shift):
    # Apply the shift to the image
    shifted_img = fourier_shift(fft2(img), shift)
    shifted_img = np.abs(ifft2(shifted_img))

    return shifted_img
```
